```python
from typing import Tuple, List

class Island():
    def __init__(self, seed: Tuple[int, int], grid: List[List[int]]):
        self.seed = seed
        self.hashset = set([seed])
        n = len(grid)
        self.n = n
        tape = [seed]
        p = 0
        while p < len(tape):
            i, j = tape[p]
            if (i+1) < n:
                v = grid[i+1][j]
                if v == 1 and (i+1, j) not in self.hashset:
                    self.hashset.add((i+1, j))
                    tape.append((i+1, j))
            if (i - 1) > -1:
                v = grid[i-1][j]
                if v == 1 and (i-1, j) not in self.hashset:
                    self.hashset.add((i-1, j))
                    tape.append((i-1, j))
            if (j+1) < n:
                v = grid[i][j+1]
                if v == 1 and (i, j+1) not in self.hashset:
                    self.hashset.add((i, j+1))
                    tape.append((i, j+1))
            if (j-1) > -1:
                v = grid[i][j-1]
                if v == 1 and (i, j-1) not in self.hashset:
                    self.hashset.add((i, j-1))
                    tape.append((i, j-1))
            p += 1
            
    def add_one_depth(self):
        tape = [x for x in self.hashset]
        p = 0
        while p < len(tape):
            i, j = tape[p]
            if (i+1) < self.n:
                self.hashset.add((i+1, j))
            if (i - 1) > -1:
                self.hashset.add((i-1, j))
            if (j+1) < self.n:
                self.hashset.add((i, j+1))
            if (j-1) > -1:
                self.hashset.add((i, j-1))
            p += 1

def check_overlap(island1, island2):
    return len(island1.hashset & island2.hashset) > 0

class Solution:
    def shortestBridge(self, grid: List[List[int]]) -> int:
        n = len(grid)
        island1 = None
        island2 = None
        for i in range(n):
            for j in range(n):
                if grid[i][j] == 1 and (island1 is None):
                    island1 = Island((i, j), grid)
                elif (
                    grid[i][j] == 1 
                    and (island1 is not None)
                    and ((i, j) not in island1.hashset)
                    and (island2 is None)):
                    island2 = Island((i, j), grid)
        counter = 0
        while not check_overlap(island1, island2):
            island1.add_one_depth()
            counter += 1
        return counter - 1
    
grid = [[0,1],[1,0]]
Solution().shortestBridge(grid)
```

```python
def _rec(nums, lp, rp):
    if rp - lp < 4:
        p = lp
        cond = True
        ret = -1
        while p <= rp and cond:
            if (nums[p] > nums[p-1]) and (nums[p] > nums[p+1]):
                cond = False
                ret = p
            p += 1
    else:   
        x0 = nums[lp]
        p1 = lp + (rp - lp) // 3
        p2 = lp + 2 * (rp - lp) // 3
        x1 = nums[p1]
        x2 = nums[p2]
        x3 = nums[rp]
        if (x1 <= x2) and (x2 >= x3):
            ret = _rec(nums, p1, rp)
        elif (x0 <= x1) and (x1 >= x2):
            ret = _rec(nums, lp, p2)
        else:
            if (nums[rp] > nums[rp-1]) and (nums[rp] > nums[rp+1]):
                ret = rp
            elif (nums[lp] > nums[lp+1]) and (nums[lp] > nums[lp-1]):
                ret = lp
            else:
                ret = _rec(nums, lp+1, rp-1)  
    return ret

class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        # O(n) is trivial
        # find O(logn)
        # check endpoints for peaks first, then check middle
        n = len(nums)
        if n > 1:
            if nums[0] > nums[1]:
                ret = 0
            elif nums[-1] > nums[-2]:
                ret = n - 1
            else:
                ret = _rec(nums, 1, n-2)
        else:
            ret = 0
        return ret
    
nums = [1,2,3,1]
Solution().findPeakElement(nums)
```

```python
from typing import List
class Heap():
    def __init__(self, heap:List[object]=[], max_size:int=None):
        self.max_size = max_size
        self.heap = (sorted(heap) if max_size is None
                     else sorted(heap[:max_size]))    
    
    @staticmethod
    def _fl(i):
        return 2*i + 1
    
    @staticmethod
    def _fr(i):
        return 2*i + 2
    
    @staticmethod
    def _fp(i):
        if i > 0:
            ret = (i - 1)//2
        else:
            ret = 0
        return ret
    
    @staticmethod
    def _fs(i):
        if i == 0:
            ret = 0
        elif i % 2 == 0:
            ret = i - 1
        else:
            ret = i + 1
        return ret
    
    def pop(self):
        n = len(self.heap)
        if n == 0:
            ret = None
        elif n == 1:
            ret = self.heap.pop()
        else:
            root = self.heap[0]
            self.heap[0] = self.heap.pop()
            n = n - 1
            pidx = 0
            pl = self._fl(pidx)
            pr = self._fr(pidx)
            while (
                    ((pl < n) and (self.heap[pidx] > self.heap[pl]))
                    or ((pr < n) and (self.heap[pidx] > self.heap[pr]))):
                if (pr >= n):
                    minidx = pl
                else:
                    minidx, val = min([
                              (pl, self.heap[pl]),
                              (pr, self.heap[pr])],
                             key=lambda x: x[1])
                swtch = self.heap[minidx]
                self.heap[minidx] = self.heap[pidx]
                self.heap[pidx] = swtch
                pidx = minidx
                pl = self._fl(pidx)
                pr = self._fr(pidx)
            ret = root
        return ret
    
    def insert(self, val):
        n = len(self.heap)
        if (self.max_size is not None) and (n >= self.max_size):
            ret = None
        else:
            self.heap.append(val)
            pidx = n
            pp = self._fp(pidx)
            ps = self._fs(pidx)
            while (self.heap[pidx]  < self.heap[pp]):
                lst = [(pidx, self.heap[pidx]), 
                       (pp, self.heap[pp])]
                if ps < len(self.heap) - 1:
                    lst.append((ps, self.heap[ps]))
                minidx, val = min(lst, key=lambda x: x[1])
                swtch = self.heap[minidx]
                self.heap[minidx] = self.heap[self._fp(pidx)]
                self.heap[self._fp(pidx)] = swtch
                pidx = pp
                pp = self._fp(pidx)
                ps = self._fs(pidx)
            ret = None
        return ret
```

```python
import math
class Heap3():
    _N = 3
    def __init__(self):
        self.heap = []
        self.prod = -1
    def insert(self, val):
        n = len(self.heap)
        if n < self._N:
            if n == 0:
                self.heap.append(val)
            elif val > self.heap[0]:
                self.heap.insert(1, val)
            else:
                self.heap.insert(0, val)
            if n + 1 == self._N:
                self.prod = math.prod(self.heap)
        elif val > self.heap[0]:
            minidx, _ = min([
                (0, val),
                (1, self.heap[1]),
                (2, self.heap[2])],
                key=lambda x: x[1])
            if minidx == 0:
                self.heap[0] = val
            else:
                swtch = self.heap[minidx]
                self.heap[minidx] = val
                self.heap[0] = swtch
            self.prod = math.prod(self.heap)
        return self.prod

def solution(ar):
    heap = Heap3()
    return [heap.insert(x) for x in ar]
```

```python
def _check_word(word, idx, forward_trie, backward_trie):
    # assume word is longer of pair of words
    # forward block
    n = len(word)
    out = set()
    p0 = n
    while p0 >= 0:
        sub = word[:p0]
        cut = word[p0:]       
        if len(sub) <= len(cut):
            j = backward_trie.search(sub)
            if (j != -1) and (idx != j):
                if (cut == cut[::-1]):
                    out.add((idx, j))
        elif (cut == cut[::-1]):
            j = backward_trie.search(sub)
            if (j != -1) and (idx != j):
                out.add((idx, j))
        p0 -= 1
    # backward block
    p0 = 0
    while p0 <= n:
        sub = word[p0:]
        sub = sub[::-1]
        cut = word[:p0]
        if len(sub) <= len(cut):
            j = forward_trie.search(sub)
            if (j != -1) and (idx != j):
                if (cut == cut[::-1]):
                    out.add((j, idx))
        elif (cut == cut[::-1]):
            j = forward_trie.search(sub)
            if (j != -1) and (idx != j):
                out.add((j, idx))        
        p0 += 1
    return out
        
class Node:
    def __init__(self, key):
        self.key = key
        self.children = dict()
        
class Trie:
    def __init__(self):
        self.root = Node(None)
        
    def search(self, word):
        ptr = self.root
        n = len(word)
        p0 = 0
        cond = True
        if n == 0:
            if "*" in ptr.children:
                ret = ptr.children["*"]
            else:
                ret = -1
        while (p0 < n) and cond:
            c = word[p0]
            if c in ptr.children:
                ptr = ptr.children[c]
            else:
                ret = -1
                cond = False
            p0 += 1
            if (p0 == n) and cond:
                if "*" in ptr.children:
                    ret = ptr.children["*"]
                else:
                    ret = -1
        return ret
    
    def insert(self, word, idx):
        ptr = self.root
        n = len(word)
        p0 = 0
        if n == 0:
            ptr.children["*"] = idx#Node(idx)
        while (p0 < n):
            c = word[p0]
            if c in ptr.children:
                ptr = ptr.children[c]
            else:
                ptr.children[c] = Node(c)
                ptr = ptr.children[c]    
            p0 += 1
            if (p0 == n):
                ptr.children["*"] = idx#Node(idx)


class Solution:
    def palindromePairs(self, words: List[str]) -> List[List[int]]:
        forward_trie = Trie()
        backward_trie = Trie()
        
        for idx, word in enumerate(words):
            forward_trie.insert(word, idx)
            backward_trie.insert(word[::-1], idx)
        out = set()
        for idx, word in enumerate(words):
            out = (out 
                  | _check_word(word, idx, forward_trie, backward_trie))
        return out
```

```python
class Solution:
    def palindromePairs(self, words: List[str]) -> List[List[int]]:
        wmap, ans = {}, []
        for i in range(len(words)):
            wmap[words[i]] = i
        for i in range(len(words)):
            if words[i] == "":
                for j in range(len(words)):
                    w = words[j]
                    if w == w[::-1] and j != i:
                        ans.append([i, j])
                        ans.append([j, i])
                continue
            bw = words[i][::-1]
            if bw in wmap:
                res = wmap[bw]
                if res != i: ans.append([i, res])
            for j in range(1, len(bw)):
                t = bw[:j]
                if t == t[::-1] and bw[j:] in wmap:
                    ans.append([i, wmap[bw[j:]]])
                t = bw[j:len(bw)]
                if t == t[::-1] and bw[:j] in wmap:
                    ans.append([wmap[bw[:j]], i])
        return ans

```

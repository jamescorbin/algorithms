```python
def triangle_nums_boundary(tri, center, n):
    #triangle_nums = lambda x: x * (x + 1) // 2
    lf_over = (tri - (center + 1) 
               if (tri - (center + 1)) >= 0 else 0) 
    rt_over = (tri - (n - center) 
               if (tri - (n - center)) >= 0 else 0)
    dl = lf_over * (lf_over + 1) // 2
    dr = rt_over * (rt_over + 1) // 2
    ret = tri * tri - dl - dr
    return ret


class Solution:
    def maxValue(self, n: int, index: int, maxSum: int) -> int:
        base = [1 for i in range(n)]
        sm = n
        diff = maxSum - n       
        tri = 1
        while diff - triangle_nums_boundary(tri, index, n) >= 0:
            tri += 1
        return tri
            
n = 6
index = 1
maxSum = 10
Solution().maxValue(n, index, maxSum)
```

```python
class NTree:
    def __init__(self, key, val=None, children=dict()):
        self.key = key
        self.val = val
        self.children = children
        
    
def _standardize(s):
    t = s.split(".")
    if len(t) == 3:
        ret = (int(t[0]), int(t[1]), int(t[2]))
    elif len(t) == 2:
        ret = (int(t[0]), int(t[1]), -1)
    else:
        ret = (int(t[0]), -1, -1)
    return ret

def _reverse(tup):
    a = [str(d) for d in tup if d != -1]
    a = ".".join(a)
    return a
    
    
def version_numbers(ar):
    root = NTree(None)
    stand = [_standardize(s) for s in ar]
    while len(stand) > 0:
        nxt = stand.pop()
        majors = root.children
        if nxt[0] not in majors:
            new_root = NTree(nxt[0])
            root.children[nxt[0]] = new_root
        major_root = root.children[nxt[0]]
        minors = major_root.children
        if nxt[1] not in minors:
            new_root = NTree(nxt[1])
            major_root.children[nxt[1]] = new_root
        minor_root = major_root.children[nxt[1]]
        version = NTree(nxt[2], val=nxt)
        minor_root.children[nxt[2]] = version
    # unfold
    ret = []
    while root:
        major_children = list(reversed(sorted(root.children.keys())))
        while len(major_children) > 0:
            major = major_children.pop()
            minor_children = list(reversed(
                                sorted(major.children.keys())))
            while len(minor_children) > 0:
                minor = minor_children.pop()
                versions = list(reverse(sorted(
                                        minor.children.keys())))
                while len(versions) > 0:
                    node = versions.pop()
                    ret.append(_reverse(node.val))      
    return ret
```
